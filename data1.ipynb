```python
# Basic cleaning
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df['Micro_Market'].value_counts()
```

```python
# Normalize text columns
df.loc[:,'Micro_Market']=df['Micro_Market'].str.title().str.strip()
df['Micro_Market'].value_counts()
```

```python
df['Configuration'].value_counts()
```

```python
df.loc[:,'Configuration'] = df['Configuration'].str.strip().str.upper()
df['Configuration'].value_counts()
```

```python
df.loc[:,'Ticket_Price_Cr'] = df['Ticket_Price_Cr'].replace({'₹': '', ' Cr': ''}, regex=True)
df.loc[:,'Ticket_Price_Cr'] = pd.to_numeric(df['Ticket_Price_Cr'], errors='coerce')
```

```python
numeric_cols= ['Unit_Size_Sqft','Ticket_Price_Cr','Amenity_Score']


import seaborn as sns
for col in numeric_cols:
    plt.figure(figsize=(6,3))
    sns.histplot(df[col], bins=50, kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
from scipy.stats import skew
skewness = df[numeric_cols].skew()
print(skewness)
```

```python
mean_col= ['Unit_Size_Sqft','Amenity_Score']
median_col= ['Ticket_Price_Cr']
  
#Handling missing values -numeric columns
from sklearn.impute import SimpleImputer 

imputer = SimpleImputer(strategy='mean')
df.loc[:,mean_col] = imputer.fit_transform(df[mean_col])

imputer = SimpleImputer(strategy='median')
df.loc[:,median_col]= imputer.fit_transform(df[median_col])

#Handling missing values -text columns
df.loc[:,'Buyer_Comments'] = df['Buyer_Comments'].fillna("No Comments")
df.isnull().sum().to_list()
```

```python
#Correct Price per Sqft
df.loc[:,"Price_per_Sqft"] = df["Ticket_Price_Cr"] * 10000000 / df["Unit_Size_Sqft"]
```

```python
df['Purchase_Quarter'].value_counts()
```

```python
# Extract Quarter Number from Purchase_Quarter (e.g., 2024-06-30 → Q2)
df.loc[:,'Quarter_Number'] = pd.to_datetime(df['Purchase_Quarter']).dt.quarter
df
```

```python
df['Transaction_Type'].value_counts()
```

```python
# Create Booking_Flag from Transaction_Type
# Example: Primary = "Booked", Secondary = "Not Booked"
df['Booking_Flag'] = df['Transaction_Type'].map({'Primary': "Booked", 'Secondary': "Not Booked"})
```

```python
#Renaming column name as per use cases
df = df.rename(columns={"Property_ID": "Project_ID","Developer_Name": "Builder"})
```

```python
#Total data count in datasets
query0="""
SELECT COUNT(*) AS total_data
FROM luxury_housing_sales_analysis;
"""
res_0=pd.read_sql(query0,engine)
res_0
```

```python
#Booking status count
query1 = """
SELECT "Booking_Flag", COUNT(*) AS count
FROM luxury_housing_sales_analysis
GROUP BY "Booking_Flag"
ORDER BY "Booking_Flag" ASC;
"""
res_1 = pd.read_sql(query1,engine)
res_1
```

```python
# Average ticket priceby builder
query2 = """
SELECT "Builder", AVG("Ticket_Price_Cr") AS avg_ticket_price
    FROM luxury_housing_sales_analysis
    GROUP BY "Builder"
    ORDER BY "avg_ticket_price" DESC;
   
"""
res_2=pd.read_sql(query2,engine)
res_2
```

```python
#micro_market in average price
import math

query3= """SELECT "Micro_Market",
       AVG("Ticket_Price_Cr") AS avg_price,
       SUM("Ticket_Price_Cr") AS total_value
       FROM luxury_housing_sales_analysis
       GROUP BY "Micro_Market"
       ORDER BY "total_value" DESC
       LIMIT 10;"""
res_3=pd.read_sql(query3,engine)
res_3
```
